```
def prism(v, color=None):
    plot = line3d([v[i] for i in [0,1,2,0,3,4,1,4,5,2,5,3]], thickness=5, color=color) \
           + polygons3d([[0,1,2],[3,5,4],[0,3,1],[3,4,1],[1,4,2],[4,5,2],[2,5,0],[5,3,0]], v, opacity=0.5, color=color)
    
    return plot

def prismFaces(v, color=None):
    plot = polygons3d([[0,1,2],[3,5,4],[0,3,1],[3,4,1],[1,4,2],[4,5,2],[2,5,0],[5,3,0]], v, opacity=0.5, color=color)
    
    return plot
```

```
v = [vector((0,0,0)),
     vector((1,1,1)), 
     vector((-1,1,1)),
     vector((-1,-1,1)),
     vector((1,-1,1)),
     vector((1,1,-1)),
     vector((-1,1,-1)),
     vector((-1,-1,-1)),
     vector((1,-1,-1))]
```

```
cube = line3d([v[l] for l in [1,2,3,4,1,5,6,2,6,7,3,7,8,4,8,5]],
             color='red')
diagonals = line3d([v[l] for l in [1,3,6,8]], color='blue') + line3d([v[l] for l in [2,5,4,7]], color='green') 
innerTriangles = polygons3d([[0,1,2],[0,2,3],[0,3,4]], [v[l] for l in [0,1,3,6,8]], color='blue', opacity=0.5) \
                 + polygons3d([[0,1,2],[0,2,3],[0,3,4]], [v[l] for l in [0,2,5,4,7]], color='green', opacity=0.5)
    
labels = sum([text3d(l, p*1.1) for l, p in zip(vname, v)])

plot = cube + diagonals + 0 + labels
show(plot, viewer='threejs', frame=False, online=True)
```
